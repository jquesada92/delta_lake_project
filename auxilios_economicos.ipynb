```python
import requests
from pandas import read_html,concat
from datetime import datetime as dt
import unicodedata
inicio_proceso = dt.now()
main_folder = 'laprensa'
schema_folder = 'auxilios'
raw_data = f'{main_folder}/{schema_folder}/raw_data'
df = concat(read_html(requests.get('https://cdn.corprensa.com/auxilios/').content)).set_index('#')

def limpieza_header(x:str)->str:
    nfkd_form = unicodedata.normalize('NFKD', x)
    only_ascii = nfkd_form.encode('ASCII', 'ignore')
    name = only_ascii.decode('utf-8').replace(' ','_').lower().replace('#','no').replace('ano','periodo')
    return name
df.columns =  list(map(limpieza_header,df.columns))


df.to_csv(f"{raw_data}/{dt.now().strftime('%Y%m%d')}_auxilios_laprensa.csv")
df['monto'] = df.monto.str.replace('B/.|,','').astype(float)
```

```python
df.groupby('a_favor_de').agg({'monto':['sum','count']}).sort_values(('monto','sum'),ascending=False).head(20)
```
